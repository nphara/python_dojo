# Matplotlib animation

Jupyter Notebook for <s>  in-depth  </s> exploration of animation features in Matplotlib

```python
# libs: plot
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# libs: generate and transform data

import numpy as np
import pandas as pd
from random import randint
```

## Example: Sine wave

```python
%matplotlib widget
## [exemplo: seno] 
## Fonte : https://matplotlib.org/2.0.2/examples/animation/simple_anim.html
fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def animate(i):
    line.set_ydata(np.sin(x + i/10.0))  # update the data
    return line,


# Init only required for blitting to give a clean slate.
def init():
    line.set_ydata(np.ma.array(x, mask=True))
    return line,

ani = animation.FuncAnimation(fig, animate, np.arange(1, 200), init_func=init,
                              interval=25, blit=True)
```

```python
ani.event_source.stop()
```

## Example: Random data

```python

%matplotlib widget
## https://pythonforundergradengineers.com/live-plotting-with-matplotlib.html

# empty lists for x and y data
x2 = []
y2 = []

# disable automatic plot after plt.figure()
plt.ioff()

# figure and axes objects
fig2, ax2 = plt.subplots()

def animate2(i):
    pt = randint(-5,6)
    x2.append(i)
    y2.append(pt)
    ax2.clear()
    ax2.plot(x2,y2)
    ax2.set_xlim([0,20])
    ax2.set_ylim([-6,7])

ani2 = animation.FuncAnimation(fig2, animate2, frames=20, interval=500, repeat=False)

# show plot
plt.show()
```

```python
# Stops plot
ani2.event_source.stop()
```

## Example: Imported data from file

This example takes values from `data.txt` file. Add more integers and save to see the plot changing.

```python
plt.ioff()

data = [3, 6, 2, 1, 8]
fig3, ax3 = plt.subplots()
def animate3(i):
    with open('data.txt', 'r') as f:
        for line in f:
            data.append(int(line.strip()))
    ax3.clear()
    ax3.plot(data[-5:])
    
```

```python
%matplotlib widget
ani3 = animation.FuncAnimation(fig3, animate3, interval=1000, cache_frame_data=False)
plt.show()
```

```python
ani3.event_source.stop()
```
